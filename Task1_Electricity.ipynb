To know the null values in given dataset

```python
# Assuming df is your DataFrame
columns_with_nulls = ['Legislative District', 'County', 'City', 'Postal Code', 'Vehicle Location', 'Electric Utility', '2020 Census Tract']

data = data.dropna(subset=columns_with_nulls)

# Verify that null values have been removed
print("Null counts after removal:")
for column in columns_with_nulls:
    print(f"{column}: {data[column].isnull().sum()}")
```

To count the electric vehicles in the given dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named df
plt.figure(figsize=(10, 6))
sns.histplot(data['Electric Range'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.show()
```

Graph which indicates the electric range in the given dataset

```python
plt.figure(figsize=(12, 6))
sns.countplot(x=data['Legislative District'], color='skyblue')
plt.title('Count of Vehicles in Each Legislative District')
plt.xlabel('Legislative District')
plt.ylabel('Count')
plt.show()
```

Graph which indicates the Count of vehicles in the given dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Producer' is the column representing the producers
plt.figure(figsize=(12, 6))
sns.countplot(x='Make', data=data, palette='viridis', order=data['Make'].value_counts().index)
plt.title('Number of Electric Vehicles by Makers')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

This graph indiactes the relation between the vehicles and its different models

```python
import matplotlib.pyplot as plt

# Assuming 'Vehicle Type' is the column representing BEV and PHEV
vehicle_type_counts = data['Make'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Distribution of BEV vs PHEV')
plt.show()
```

This piechart shows the distribution of BEV and PHEV vehicles in the given dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Maker' is the column representing the vehicle makers
top_makers = data['Make'].value_counts().nlargest(20).index

# Filter the DataFrame for the top 20 makers
data_top_makers = data[data['Make'].isin(top_makers)]

# Plotting the maximum electric range for each maker
plt.figure(figsize=(12, 8))
sns.barplot(x='Electric Range', y='Make', data=data_top_makers, ci=None, palette='viridis')
plt.title('Maximum Electric Range of Top 20 Makers')
plt.xlabel('Electric Range')
plt.ylabel('Make')
plt.show()
```

This Graph clearly indicates maximum electric range of Top 20 Makers of every type of vehicles

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with numeric variables
numeric_variables = data[['Model Year', 'Electric Range', 'Base MSRP', 'Legislative District', '2020 Census Tract']]

# Calculate the correlation matrix
correlation_matrix = numeric_variables.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

To understand the linear relationship between different variables in a dataset. Each entry in the matrix represents the correlation coefficient between two variables. The correlation coefficient is a statistical measure that quantifies the strength and direction of a linear relationship between two variables. It ranges from -1 to 1:

1 indicates a perfect positive linear relationship. -1 indicates a perfect negative linear relationship. 0 indicates no linear relationship.

```python
sns.pairplot(data[['Electric Range', 'Base MSRP', 'Legislative District', '2020 Census Tract']])
plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
plt.show()
```

This graph indicates the pair plots to visualize relationships between multiple numeric variables. This is particularly useful for identifying potential correlations

```python
correlation_matrix = data[['Electric Range', 'Base MSRP', 'Legislative District', '2020 Census Tract']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
```

A heatmap to visualize the correlation between numeric variables. This can help identify strong correlations and guide further analysis.
