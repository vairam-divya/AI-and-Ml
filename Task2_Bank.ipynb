**Univariate Analysis for Numerical Variables**

Uni means single and variate means variable. Hence univariate means the analysis of a single variable. In univariate analysis, we are trying to analyse the pattern present in a single variable


**Histogram:** Histogram is used to plot the graphical representation of a numerical variable.

```python
import matplotlib.pyplot as plt
plt.hist(data["balance"])           # Histogram plot using matplotlib
plt.xlabel("balance")
plt.ylabel("Count")
```

From the above graph,we can say that Zero balance customers are more in this bank compared to other balance customers.

```python
import seaborn as sns
sns.histplot(data, x='balance')
```

Now for the same variable,we plotted the histogram plot for the balance entity.We can clearly see the difference between zero balance customers and other balance customers.

```python
plt.boxplot(data["balance"])      # Box plot using matplotlib
```

From the above graph,we can say that 75% of customers in bank are having balance in the range of 0-1,25,000 and also 25% of customers are in the range of 1,50,00-2,00,00.

```python
sns.boxplot(data["balance"])                     # Box plot using seaborn
```

Here also the same method which we implemented in above cell,but here we are using box plot using seaborn.

**Violin plot = Box plot + Histogram plot**

```python
sns.violinplot(data["balance"])        # Violin plot using seaborn
```

Here we can clearly say that zero balance customers are more comapared to other customers so that's why there is a more difference in the shape of the plot.

```python
import seaborn as sns
sns.violinplot(data["age"])
```

Here we can clearly say that 30-40 aged customers are more comapared to other customers so that's why there is a more difference in the shape of the plot.

**Bivariate Analysis of Numerical Variables**

**Bivariate means the analysis of two variables. Using bivariate analysis we can find how well the variables are correlated.**

Scatterplot: Scatterplot uses dots to represent the relationship between two numeric variables.

```python
sns.countplot(x='gender', hue='churn', data=data)
plt.title('Churn by Gender')
plt.show()
```

From the above graph we can say that,male customers have more count of churn compared to the female customers in the bank.It shows the relationship between the churn and gender in dataset.

**Count plot: Using the hue parameter we can analyse two categorical variables in a count plot.**

```python
sns.countplot(x='churn', data=data)
plt.title('Distribution of Churn')
plt.show()
```

Here the count plot indicates the distribution of churn among the bank customers.

```python
plt.scatter(data["balance"], data["age"])
plt.title("Scatter Plot of balance vs age")
plt.xlabel("balance")
plt.ylabel("age")
plt.show()
```

Here we can say that,more number of 30-50 aged customer are having more balance compared to other customers


```python
import matplotlib.pyplot as plt
sns.boxplot(x='gender', y='balance', data=data)
plt.title('Balance Distribution by Gender')
plt.show()
```

From the above box plot we can say that both male and female has equal range of balance in range of 0-1,25,000.

```python
ct = pd.crosstab(data['country'], data['active_member'])
print(ct)
sns.heatmap(ct, annot=True, cmap='Blues')
plt.title('Active Membership by Geography')
plt.show()
```

From above we can say that active menbers of bank are more from the france country compared to other countries.

```python
categorical_features = data.select_dtypes(include=['object']).columns
for feature in categorical_features:
    sns.countplot(x=feature, hue='churn', data=data)
    plt.title(f'Churn by {feature}')
    plt.show()
```

From the above graphs,
France country has more churn compared to other countries.
Among male and female,male customers are having more churn compared to female customers.


```python
sns.violinplot(x='country', y='balance', data=data)
plt.title('Balance Distribution by country(Violin Plot)')
plt.show()
```

From above violon plot,germany customers have more balance compared to other countries in the range of 1,00,000-1,50,000 while in france and spain coutries people have more zero balance customers.

```python
numerical_columns = data.select_dtypes(include=['number']).columns.tolist()
categorical_columns = data.select_dtypes(exclude=['number']).columns.tolist()

print("Numerical Columns:", numerical_columns)
print("Categorical Columns:", categorical_columns)
```

```python
import matplotlib.pyplot as plt
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Pie Chart for {column}')
    plt.ylabel('')
    plt.show()
class_counts = data['churn'].value_counts()
fig = plt.figure(figsize=(5,5))
plt.pie(class_counts, labels=['not churned','churned'],explode=(0,0.1),shadow=True, autopct="%1.1f%%", startangle=180)
plt.title('No. of Customers Churned v/s Not Churned')
plt.show()

class_counts = data['active_member'].value_counts()
fig = plt.figure(figsize=(5,5))
plt.pie(class_counts, labels=['not active_member','active_member'],explode=(0,0.1),shadow=True, autopct="%1.1f%%", startangle=180)
plt.title('No. of Customers Churned v/s Not Churned')
plt.show()
```

From above pie charts,
1)France country has more customers compared to other country customers.

2)Male customers are more comapred to female customers.

3)Not churned customers are more compared to churned customers.

4)Now again churned Vs Unchurned customers but among active members.
