**Univariate Analysis for Categorical Variables:**

Univariate analysis of categorical variables can be performed using the following methods.

Count plot: The count plot is similar to the histogram. But unlike histogram count plot represents frequency distribution of categorical data.

```python
import seaborn as sns
sns.countplot(data, x="Electric Vehicle Type")
```

From the above graph,we can clearly say that BEV vehicles are high comapared to PHEV vehicles

```python
# Pie chart using matplotlib
import matplotlib.pyplot as plt
dependents = data['Electric Vehicle Type'].value_counts()
plt.pie(dependents.values, labels=dependents.index, autopct='%1.1f%%')
plt.title("Electric Vehicle Type Distribution")
```

This pie chart indicates the different types of electric vehicles type distribution in the given dataset.


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['Electric Range'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.show()
```

Above Graph indicates the electric range in the given dataset,which indicates that Electric range is high at 3000 and slowly its decreasing.

```python
plt.figure(figsize=(12, 6))
sns.countplot(x=data['Legislative District'], color='skyblue')
plt.title('Count of Vehicles in Each Legislative District')
plt.xlabel('Legislative District')
plt.ylabel('Count')
plt.show()
```

The above garph counts the different types of vehicles in each district,which indicates the 41st district has highest count of vehicles.

**Bivariate Analysis**

Bivariate means the analysis of two variables. Using bivariate analysis we can find how well the variables are correlated.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(x='Make', data=data, palette='viridis', order=data['Make'].value_counts().index)
plt.title('Number of Electric Vehicles by Makers')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

In the above graph,We counted the number of elctric vehicles by makers so there are two variables are used which are Make and Electric vehicles.

```python
import matplotlib.pyplot as plt

vehicle_type_counts = data['Make'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Distribution of BEV vs PHEV')
plt.show()
```

This piechart shows the distribution of BEV and PHEV vehicles in the given dataset,which indicates that tesla has more vehicles comapred to other types.

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_makers = data['Make'].value_counts().nlargest(20).index

data_top_makers = data[data['Make'].isin(top_makers)]

plt.figure(figsize=(12, 8))
sns.barplot(x='Electric Range', y='Make', data=data_top_makers, ci=None, palette='viridis')
plt.title('Maximum Electric Range of Top 20 Makers')
plt.xlabel('Electric Range')
plt.ylabel('Make')
plt.show()
```

The above graph clearly indicates that Chetvrolet Makers are high compared to other makers.And it also shows the top 20 makers in given dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.violinplot(x='Electric Vehicle Type', y='Electric Range', data=data)
plt.title('Violin Plot of Electric Range by Your Column')
plt.show()
```

Here we can clearly say that BEV are having more range 0 and PHEV vehicles are in range of 25-50 so that's why there is a change in shape of violin plot.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Electric Vehicle Type', y='Electric Range', data=data)
plt.title('Box Plot of Electric Range by Your Column')
plt.show()
```

From the above box plot we can say that 75% of BEV vehicles are in the range of 0-200 and 25% are in the range of 200-300 whereas in PHEV vehicles,75% are in the range of 25-50 and remaining 25% in range of 50-75.

```python
import matplotlib.pyplot as plt
plt.scatter(data["Electric Vehicle Type"], data["Model Year"])
plt.title("Scatter Plot of Electric Vehicle Type vs Model Year")
plt.xlabel("Electric Vehicle Type")
plt.ylabel("Model Year")
plt.show()
```

From the above scatter plot,both BEV and PHEV vehicles are having equal models in the year 2025.But PHEV vehilces started in 2010 whereas BEV vehicles are from 2000.
